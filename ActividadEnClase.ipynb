```python
from ipywidgets import widgets
from IPython.display import display

slider = widgets.IntSlider(
    value = 7,
    min = 0,
    max = 10,
    step = 1,
    decription = 'Value: ',
    disabled = False,
    continuos_update = False,
    orientation = 'horizontal',
    readout = True,
    readout_format = 'd'
)

display(slider)

def on_value_change(change):
  print(f"Slider value changed to: {change['new']}")

slider.observe(on_value_change, names = 'value')
```

```python
from ipywidgets import widgets
from IPython.display import display

text_input = widgets.Text(
    value = 'Hello World!',
    placeholder = 'Type something',
    description = 'Text: ',
    disabled = False
)

output_area = widgets.Output()

display(text_input)
display(output_area)

def on_text_change(change):
  with output_area:
    output_area.clear_output()
    print(f"You typed: {change['new']}")

text_input.observe(on_text_change, names = 'value')
```

```python
from ipywidgets import widgets
from IPython.display import display

dropdown = widgets.Dropdown(
    options = ['Phyton', 'Arturo', 'C++', 'Javascript', 'Julia'],
    value = 'Phyton',
    description = 'Language: ',
    disbabled = False,
)

display(dropdown)

def on_dropdown_change(change):
  print(f"Selected language: {change['new']}")

dropdown.observe(on_dropdown_change, names = 'value')
```

```python
from ipywidgets import widgets
from IPython.display import display

button1 = widgets.Button(description = "Button 1")
button2 = widgets.Button(description= "Button 2")
output = widgets.Output()

display(button1)
display(button2)
display(output)

def on_button1_clicked(b):
  with output:
    output.clear_output()
    print("Button 1 clicked!")

def on_button2_clicked(b):
  with output:
    output.clear_output()
    print("Button 2 clicked!")

button1.on_click(on_button1_clicked)
button2.on_click(on_button2_clicked)
```

```python
import os
def list_folders(start_path= '/', output_file = 'folders_list.txt'):
  excluded_dirs = {'/bin', '/boot', '/dev', '/etc', '/lib', '/lib64', '/proc', '/run', '/sbin', '/sys', '/tmp', '/var'}

  with open(output_file, 'w') as f:
    for root, dirs, _ in os.walk(start_path):
      dirs[:] = [d for d in dirs if os.path.join(root, d) not in excluded_dirs]
      for directory in dirs:
        folder_path = os.path.join(root, directory)
        print(folder_path)
        f.write(folder_path + '\n')

if __name__ == "__main__":
  list_folders('/')
```
