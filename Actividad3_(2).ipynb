```python
from IPython.display import Image, display

class EmpleadoUniversitario:
  def __init__(self, nombre, departamento, salario, foto):
    self._nombre = nombre
    self._departamento = departamento
    self._salario = salario
    self._foto = foto

  @property
  def nombre(self):
    return self._nombre
  @property
  def departamento(self):
    return self._departamento
  @property
  def salario(self):
    return self._salario
  @property
  def foto(self):
    return self._foto

  @nombre.setter
  def nombre(self, nombre):
    self._nombre = nombre
  @departamento.setter
  def departamento(self, departamento):
    self._departamento = departamento
  @salario.setter
  def salario(self, salario):
    if salario < 0:
      print("No puede ser negativo el salario")
      return
    self._salario = salario
  @foto.setter
  def foto(self, foto):
    self._foto = foto


  def show_image(self):
    if self.foto:
      display(Image(self.foto))
    else:
      print("No image available.")

empleado1 = EmpleadoUniversitario("Carlos", "Sistemas", 1, "https://www.cnn.com/interactive/2020/08/sport/lionel-messi-career-barcelona-spt-intl/media/images/card.jpg")
print("Datos iniciales")
print(empleado1.nombre)
print(empleado1.departamento)
print(empleado1.salario)
empleado1.show_image()

empleado1.nombre = "Juan"
empleado1.departamento = "Administracion"
empleado1.salario = 5500
empleado1.foto = "https://upload.wikimedia.org/wikipedia/commons/thumb/4/47/PNG_transparency_demonstration_1.png/640px-PNG_transparency_demonstration_1.png"

print("\nDatos modificados")
print(empleado1.nombre)
print(empleado1.departamento)
print(empleado1.salario)
empleado1.show_image()

empleado1.salario = -3
print(empleado1.salario)
```
